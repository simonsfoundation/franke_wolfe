```python
import sys
if ".." not in sys.path:
    sys.path.append("..")
from franke_wolfe import fw_optimize
import fw_draw
import numpy as np
import math
```

```python
bounds = [(0, 1), (0, 4)]
(cx, cy) = (4, 2)
cx = 4
cy = 1.66666666
def objective_2d(a):
    [x, y] = a
    return (x - cx) ** 2 + (y - cy) ** 2
lm = 5
options = {}
perturb = True
runner = fw_optimize.FrankWolfe(objective_2d, bounds=bounds, options=options, iteration_limit=lm)
if perturb:
    A_ub = [[-0.1, -1]]
    b_ub = [-0.1]
    runner = fw_optimize.FrankWolfe(objective_2d, 
                                    A_ub = A_ub,
                                    b_ub = b_ub,
                                    bounds=bounds, 
                                    options=options, 
                                    iteration_limit=lm)
#runner.verbose = True
runner.vertex_ratio = 0.66
runner.blend = False
ill = fw_draw.Illustration(runner)
ill.location((cx, cy))
ill.location((0, 0))
D = ill.get_canvas()
radius = 0.01 * np.max(ill.maxes - ill.mins)
D.circle(None, cx, cy, (ill.maxes[0]-ill.mins[0])*0.05, "red")
if perturb:
    D.line(None, 0, 0.1, 1, 0, "magenta")
D = ill.draw()
D.embed()
```
